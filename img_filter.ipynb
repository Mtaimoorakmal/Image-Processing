```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path to your local image
image_path = '/content/sic 1.jpg'  # Replace with your image path

# Step 1: Read the image using OpenCV
image = cv2.imread(image_path)

# Step 2: Convert the image from BGR to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Canny edge detection
edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

# Step 4: Display the original, grayscale, and edge-detected images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
# plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
# plt.imshow(gray_image)

plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
# plt.imshow(edges)
plt.title('Edge Detected Image')
plt.axis('off')

plt.show()
```

```python
# Define the Sobel filter kernels
sobel_horizontal = np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])

sobel_vertical = np.array([[-1, -2, -1],
                         [ 0,  0,  0],
                         [ 1,  2,  1]])

# Step 1: Read the image using OpenCV
image = cv2.imread(image_path)

# Step 2: Convert the image from BGR to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply the Sobel filter manually using cv2.filter2D
edges_horizontal = cv2.filter2D(gray_image, -1, sobel_horizontal)
edges_vertical = cv2.filter2D(gray_image, -1, sobel_vertical)

# Step 4: Combine the horizontal and vertical edges
edges_combined = np.sqrt(np.square(edges_horizontal) + np.square(edges_vertical))
edges_combined = np.uint8(edges_combined)

# Step 5: Display the original and edge-detected images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges_horizontal, cmap='gray')
plt.title('Horizontal Edges')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges_vertical, cmap='gray')
plt.title('Vertical Edges')
plt.axis('off')

plt.figure(figsize=(5, 5))
plt.imshow(edges_combined, cmap='gray')
plt.title('Combined Edges')
plt.axis('off')

plt.show()
```
